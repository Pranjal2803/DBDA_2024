##Question: Imagine a scenario where an investor wants to understand the risks and returns associated with various stocks before investing in them.  Given two stocks, find out which stock is at lesser risk by looking into measures of dispersion of stocks values. Also implement the  sub task  for the same problem


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns
import warnings

## load beml data
beml_df=pd.read_csv('BEML.csv')
beml_df[0:5]
```

```python
## load the glaxo data 
print("glaxo")
glaxo_df=pd.read_csv('GLAXO.csv')
glaxo_df[0:5]
```

```python
## selecting date and time in dataframe

beml_df=beml_df[['Date', 'Close']]
glaxo_df=glaxo_df[['Date' ,'Close']]
```

```python
## change time to index 
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
## check information of beml stock
beml_df.info()
```

```python
## plotting the close data of glaxo stock
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

plt.plot(glaxo_df.Close);
plt.xlabel('Time');
plt.ylabel('Close Price');
plt.title("plotting  graph of glaxo stock")
```

```python
## plotting the close data of beml stock
plt.plot(beml_df.Close);
plt.xlabel('Time');
plt.ylabel('Close Price');
plt.title("plotting  graph of glaxo stock")
```

```python
## claculating gain 
glaxo_df['gain']=glaxo_df.Close.pct_change(periods =1)

beml_df['gain']=beml_df.Close.pct_change(periods=1)
glaxo_df.head(5)
```

```python
##ploting gain aganinst time
plt.figure(figsize =(8, 6));
plt.plot(glaxo_df.index,glaxo_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
##plotinf gain of both stocks
import warnings
warnings.filterwarnings('ignore')

sns.distplot(glaxo_df.gain ,label='Glaxo')
sns.distplot(beml_df.gain ,label='beml')
plt.xlabel('gain')
plt.ylabel('Density')
plt.legend()
```

```python
## print daily gain
print("daily gain of glaso")
print("Mean: ", round(glaxo_df.gain.mean(),4))
print("standard derivation:" ,round(glaxo_df.gain.std(),4))
      
```

```python
## print daily gain
print("daily gain of beml")
print("Mean: ", round(beml_df.gain.mean(),4))
print("standard derivation:" ,round(beml_df.gain.std(),4))
      
```

```python
## describe and  risk of stack
print("beml stock")
beml_df.gain.describe()
```

```python
## describe and  risk of stack
print("glaxo stock")
glaxo_df.gain.describe()
```

```python
## beml stock has higher risk as standard devivation of beml is 2.64% 
## Where as the standard devivation of glaxo ix 1.33%
```

3) In Python, the describe() function from the statistics module is used to compute basic statistical properties of a dataset. It provides a summary of the dataset's central tendency, dispersion, and shape. This function is particularly useful for exploratory data analysis and getting a quick overview of the data

The output will provide a DescribeResult object containing the following statistical properties:

nobs: The number of observations in the dataset.
minmax: A tuple containing the minimum and maximum values in the dataset.
mean: The arithmetic mean of the dataset.
variance: The variance of the dataset.
stddev: The standard deviation of the dataset.
This function provides a convenient way to obtain these statistics without having to compute them individually.

```python
# In the extract step, we're reading data from the CSV file named "onestop.csv" using the pd.read_csv() function from the Pandas library. 
# This function reads the data and stores it in a DataFrame named passport_df. The print(onestop_df.head(5)) statement displays the first 5 rows of 
# the DataFrame, giving us a glimpse of the data we've just extracted.
```

```python
onestop_df=pd.read_csv("onestop.csv")
onestop_df.head(5)
```

```python
#plot distribution plot
# use distplot.plt() funtion for ploting distribution of both stock  gain 
# for plotting graph use the sns.distplot() function  and set label of  both axis  use plt.xlabel() and plt.ylabel()
```

```python
sn.distplot(onestop_df['discount_0'],label="no discounts")
sn.distplot(onestop_df['discount_10'],label="10% discounts")
sn.distplot(onestop_df['discount_20'],label="20% discounts")
```

```python
#for performing one anova testing we use  stats.f_oneway() function
```

```python
stats.f_oneway(onestop_df['discount_0'],onestop_df['discount_10'],onestop_df['discount_20'])
```

##### Conclusion : 3.821500-18 <0.05 : reject H0

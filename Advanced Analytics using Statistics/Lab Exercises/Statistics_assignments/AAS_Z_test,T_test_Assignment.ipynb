# Z-test

```python
# In the extract step, we're reading data from the CSV file named "passport.csv" using the pd.read_csv() function from the Pandas library. 
# This function reads the data and stores it in a DataFrame named passport_df. The print(passport_df.head(5)) statement displays the first 5 rows of 
# the DataFrame, giving us a glimpse of the data we've just extracted.
```

```python
passport_df=pd.read_csv("passport.csv")
passport_df.head(5)
```

```python

#for performing z_test we use def z_test() function
# and needs population mean , polpulation stardard and sample data  and square root of lengthof sample data
```

```python
def z_test(pop_mean,pop_std,sample):
    z_score=(sample.mean()-pop_mean)/pop_std/math.sqrt(len(sample))
    return z_score,stats.norm.cdf(z_score)
```

```python
z_test(30,12.5,passport_df.processing_time)
```

```python
#Here the p-value is greater  than aplha value i.e 5% or 0.05 ,so claim made by null hypothesis is accepted.
```

##### Conclusion : 0.4851 > 0.05 ;Fail to reject H0(accept HO) 

# One  T-test

```python
# In the extract step, we're reading data from the CSV file named "bollywoodmovies.csv" using the pd.read_csv() function from the Pandas library. 
# This function reads the data and stores it in a DataFrame named passport_df. The print(bollywood_movies_df.head(5)) statement displays the first 5 rows of 
# the DataFrame, giving us a glimpse of the data we've just extracted.
```

```python
bollywood_movies_df=pd.read_csv("bollywoodmovies.csv")
bollywood_movies_df.head(5)
```

```python

#for performing one ttest we stats.ttest_1samp() function
#  
```

```python
stats.ttest_1samp(bollywood_movies_df.production_cost,500)
```

```python
#Here the p-value is greater  than aplha value i.e 5% or 0.05 ,so claim made by null hypothesis is accepted.
```

# Two Smaple T-Test

```python
# In the extract step, we're reading data from the CSV file named "healthdrink.xlsx" using the pd.read_excel() function from the Pandas library. 
# This function reads the data and stores it in a DataFrame named passport_df. The print(healthdrink_df.head(5)) statement displays the first 5 rows of 
# the DataFrame, giving us a glimpse of the data we've just extracted.
```

```python
healthdrink_yes_df=pd.read_excel('healthdrink.xlsx',"healthdrink_yes")
healthdrink_no_df=pd.read_excel('healthdrink.xlsx','healthdrink_no')
```

```python
#for performing one ttest we stats.ttest_ind() function
```

```python
stats.ttest_ind(healthdrink_yes_df['height_increase'],healthdrink_no_df['height_increase'])
```

```python
#Here the p-value is greater  than aplha value i.e 5% or 0.05 ,so claim made by null hypothesis is accepted.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("Stud_part1.csv")
df1.head(5)
```

```python
df2 = pd.read_csv("Stud_part2.csv")
df2.head(5)
```

```python
merg= pd.merge(df1,df2,how='inner',on='regnum')
```

```python
merg.head(5)
```

```python
merg.drop(['city_y','studyhours_y','cgpa_y'],axis=1,inplace=True)
```

```python
merg.head()
```

```python
new_name=['regnum','name','attendance','theorymarks','city','studyhours' ,'cgpa']
merg.columns=new_name
```

```python
merg
```

```python
merg.isnull().sum()
```

```python
merg.info()
```

```python
merg.describe()
```

```python
# Replace non-numeric values with NaN
merg['studyhours'] = pd.to_numeric(merg['studyhours'], errors='coerce')
merg['cgpa'] = pd.to_numeric(merg['cgpa'], errors='coerce')

# Handle missing values
merg['studyhours'].fillna(merg['studyhours'].median(), inplace=True)
merg['cgpa'].fillna(merg['cgpa'].median(), inplace=True)

# Convert to int32
merg['studyhours'] = merg['studyhours'].astype(np.int32)
merg['cgpa'] = merg['cgpa'].astype(np.int32)

print(merg)
```

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.boxplot(y=merg['studyhours'])
plt.title('Box Plot of Study Hours')

plt.subplot(1, 2, 2)
sns.boxplot(y=merg['cgpa'])
plt.title('Box Plot of CGPA')

plt.tight_layout()
plt.show()
```

```python
x=merg['theorymarks']
y=merg['cgpa']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test,y_pred)
```

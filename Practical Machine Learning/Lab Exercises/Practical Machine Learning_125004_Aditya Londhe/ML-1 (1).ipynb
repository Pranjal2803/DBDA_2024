```python
# TASK -7
import re
import pandas as pd


with open('data.txt', 'r') as file:
    text = file.read()
    
pattern = r"name\s*=\s*([a-zA-Z]+)\s*.*?marks\s*=\s*\[([0-9\s,]+)\]"
matches = re.findall(pattern, text)


df = pd.DataFrame(matches, columns=['name', 'marks'])
df['marks'] = df['marks'].apply(lambda x: list(map(int, x.split())))


df['total'] = df['marks'].apply(sum)
df['avg'] = df['marks'].apply(lambda x: sum(x) / len(x))
print(df)
```

```python
# TASK - 8
import pandas as pd

alst = {
    "name": ["hari", "john", "ramu", "elan"],
    "city": ["blr", "chn", "hyd", "mum"],
    "math": [50, 60, 80, 100],
    "sci": [56, 64, 56, 43],
    "soc": [84, 54, 75, 54]
}

df = pd.DataFrame(alst)

df['total'] = df['math'] + df['sci'] + df['soc']
df['best_subject'] = df[['math', 'sci', 'soc']].idxmax(axis=1)
print(df)
print(" ")
best_students = {}
for subject in ['math', 'sci', 'soc']:
    best_student = df.loc[df[subject].idxmax(), 'name']
    best_students[subject] = best_student
    
print("Best student in math dept =", best_students['math'])
print("Best student in sci dept =", best_students['sci'])
print("Best student in soc dept =", best_students['soc'])
```

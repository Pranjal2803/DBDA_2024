### Consider Warehouse Inventory System [1D Array]:
1. Create an array consisting of product code and corresponding quantity
2. Determine the dimensions of the inventory array
3. Given a product code search through the array to find quantity of array
4. Split the array
5. Retrieve the quantity using array index
6. Check the shape of the array
7. Reshape the inventory onto 2D Array
8. Create a copy of the array
9. Add new items to the inventory

```python
import numpy as np
product_code=np.array(["p1","p2","p3","p4"])
quantity=np.array([10,20,30,40,50,60])
print(product_code,quantity)
p1=np.array_split(product_code,2)
print(p1)
print("quantity = {}".format(quantity))
i = np.where(quantity == 30) 
print("i = {}".format(i))
print(quantity[1]+quantity[3])
print(quantity.shape)
p2=quantity.reshape(2,3)
print(p2)
p3=product_code.copy
quantity[0]=42
print(quantity)
print(p3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = np.array(["Math", "Science", "History", "English", "Arts"])
subject_scores = np.array([85, 70, 75, 90, 80])

plt.bar(subjects, subject_scores, color=["blue", "pink", "red", "green", "yellow"])
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Subject Scores')
plt.grid(True,linestyle='-')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])  # Time spent studying (in hours)
y = np.array([50, 60, 70, 80])  # Study scores (out of 100)

plt.plot(x, y, color='blue', linestyle='-')
plt.scatter(x,y,color='red')
plt.xlabel('Time Spent Studying (hrs)')
plt.ylabel('Study Scores (out of 100)')
plt.title('Study Scores vs Time Spent Studying')
plt.show()

```

```python
import pandas as pd

df_train = pd.read_csv('movie_review_train.csv')
df_test = pd.read_csv('movie_review_test.csv')
df_train.head(10)
print('test')
df_test.head(20)
```

```python
def convert(x):
    d={'Pos':1,'Neg':0}
    return d[x]

# df_train['label']=df_train["class"].apply(convert)
# df_train.head()

df_test['label'] = df_test["class"].apply(convert)
df_test.head()
```

```python
df_test= df_test.drop('class',axis=1)
df_test.head()
```

```python
#convert to x and y 
x=df_test.text
y = df_test.label
print(x.shape)
print(y.shape)
```

```python
# splitting into test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(stop_words='english')
```

```python
vect.fit(x)
```

```python
#printing vocabulary
vect.vocabulary_
```

```python
#vocab size
len(vect.vocabulary_)
dict=CountVectorizer(stop_words='english',min_df=0.03,max_df=.8)
dict.fit(x)
x_train_vocabs_dict=dict.get_feature_names_out()
len(x_train_vocabs_dict)
```

```python
#transforming the train and test datasets
x_train_transformed=vect.transform(x_train)
x_test_transformed=vect.transform(x)
```

```python
print(type(x_train_transformed))
print (x_train_transformed)
```

```python
# training the NB model and making predictions
from sklearn.naive_bayes import BernoulliNB
mnb=BernoulliNB()
#fit
mnb.fit(x_train_transformed,y_train)
#predict class
y_pred_class=mnb.predict(x_test_transformed)
#predict probabilities
y_pred_proba=mnb.predict_proba(x_test_transformed)
```

```python
#printing the overall accuracy
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class[:100])
```

```python
import pandas as pd
import tldextract

# Function to count vowels and return them
def extract_vowels(s):
    vowels = "aeiouAEIOU"
    vowel_list = [char for char in s if char in vowels]
    vowel_string = ''.join(vowel_list)
    return vowel_string, len(vowel_list)

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv("test_domains.csv", header=None)

# Step 2: Extract domain components using tldextract
extracted_domains = df[0].apply(tldextract.extract)

# Step 3: Create a new DataFrame to store the extracted components
extracted_df = pd.DataFrame(extracted_domains.tolist(), columns=['subdomain', 'domain', 'suffix'])

# Step 4: Calculate the length of each domain name
extracted_df['domain_length'] = extracted_df['domain'].apply(len)

# Step 5: Calculate vowels in each domain name
extracted_df['vowels'], extracted_df['vowel_count'] = zip(*extracted_df['domain'].apply(extract_vowels))

# Step 6: Concatenate the original DataFrame with the extracted components
result_df = pd.concat([df, extracted_df], axis=1)

# Step 7: Display the result DataFrame
print(result_df.head(5))
```

```python
from collections import defaultdict
import re

def count_vowels(word):
    vowels = "aeiou"
    return sum(1 for char in word if char.lower() in vowels)

def analyze_domains(domains):
    length_dict = {}
    word_count_dict = defaultdict(int)
    vowels_count_dict = {}
    domain_vowels_dict = {}

    for domain in domains:
        # Calculate length of domain
        length_dict[domain] = len(domain)

        # Count occurrence of each word
        words = re.findall(r'\w+', domain)
        for word in words:
            word_count_dict[word] += 1

        # Count vowels
        vowels_count_dict[domain] = count_vowels(domain)

        # Save vowels in domain
        domain_vowels_dict[domain] = [char for char in domain if char.lower() in 'aeiou']

    return length_dict, word_count_dict, vowels_count_dict, domain_vowels_dict

def main():
    domains = [
        "gllcsbqrbfch.com",
        "hkgjjokkwodt.com",
        "jjnehudfmclq.com",
        "hieffvbcjitc.com",
        "fckjmggtxchv.com",
        "gotmnoxdlmqy.com",
        "lsnuwgkjbubr.com",
        "jmtkcbjfrnwp.com",
        "bkjciirqpmrs.com",
        "hcbdceyvftuf.com",
        "kglydddmwwxv.com",
        "ejddgfnoyhnn.com",
        "oovsccucdqjj.com",
        "dkrvosdssglm.com",
        "jedsotqkhiiu.com"
    ]

    length_dict, word_count_dict, vowels_count_dict, domain_vowels_dict = analyze_domains(domains)

    print("Length of each domain:")
    for domain, length in length_dict.items():
        print(f"{domain}: {length}")

    print("\nCount of each word:")
    for word, count in word_count_dict.items():
        print(f"{word}: {count}")

    print("\nNumber of vowels in each domain:")
    for domain, vowels_count in vowels_count_dict.items():
        print(f"{domain}: {vowels_count}")

    print("\nVowels in each domain:")
    for domain, vowels in domain_vowels_dict.items():
        print(f"{domain}: {''.join(vowels)}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import re

def count_vowels(word):
    vowels = "aeiou"
    return sum(1 for char in word if char.lower() in vowels)

def analyze_domains_from_file(filename):
    with open(filename, 'r') as file:
        domains = file.readlines()

    data = []
    word_count_dict = {}
    vowels_count_dict = {}

    for domain in domains:
        domain = domain.strip()  # Remove trailing newline character
        length = len(domain)
        words = re.findall(r'\w+', domain)
        word_count = {word: words.count(word) for word in set(words)}
        vowels_count = count_vowels(domain)
        vowels = ''.join(char for char in domain if char.lower() in 'aeiou')
        data.append([domain, length, word_count, vowels_count, vowels])

        word_count_dict[domain] = word_count
        vowels_count_dict[domain] = vowels_count

    columns = ['Domain', 'Length', 'Word_Count', 'Vowels_Count', 'Vowels']
    df = pd.DataFrame(data, columns=columns)
    return df, word_count_dict, vowels_count_dict

def main():
    filename = "test_domains.txt" 
    df, word_count_dict, vowels_count_dict = analyze_domains_from_file(filename)

    print("Domain Analysis:")
    print(df)

    print("\nWord Count Dictionary:")
    for domain, word_count in word_count_dict.items():
        print(domain, ":", word_count)

    print("\nVowels Count Dictionary:")
    for domain, vowels_count in vowels_count_dict.items():
        print(domain, ":", vowels_count)

if __name__ == "__main__":
    main()
```

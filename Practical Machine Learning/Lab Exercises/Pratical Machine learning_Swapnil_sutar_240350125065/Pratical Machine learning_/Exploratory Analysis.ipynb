```python
# Load Input data
import numpy as np
import pandas as pd

fitness = pd.read_excel("C:\\Users\\Training\\Data Science using SAS and Python\\Data\\fitness.xlsx")
fitness.head()
```

```python
# Get the Scatter Plot of Oxygen_Consumption with other predictors
import matplotlib.pyplot as plt
fitness.plot.scatter(x = "Run_Pulse", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "Rest_Pulse", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "Maximum_Pulse", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "Performance", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "RunTime", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "Age", y = "Oxygen_Consumption")
fitness.plot.scatter(x = "Weight", y = "Oxygen_Consumption")
plt.show()
```

```python
# Get the Pearson Correlation coeeficient of Oxygen_Consumption vs. Other Predictor variables

corr_df = fitness.corr(method = 'pearson')
corr_df = pd.DataFrame(corr_df.loc["Oxygen_Consumption"])
corr_df
```

# Variable Clustering

```python
import numpy as np
import pandas as pd

# Load the Input Data
clus_df = pd.read_excel("./var_clustering.xlsx")
clus_df.head()
```

```python
# Variable Clustering
from varclushi import VarClusHi

# Drop the Target variable. We don't need it while clustering variables
clus_df.drop('Y', axis = 1, inplace=True)

# Clustering Variables
var_clus_model = VarClusHi(clus_df, maxeigval2 = .7, maxclus = None)
var_clus_model.varclus()
```

```python
# var_clus_model.info
```

```python
# Variable Selection from each cluster based on R_Sq ratio
# RS_RATO = 1 - RS_Own_Cluster / 1 - RS_Next_Closest_Cluster
# Select one variable from each cluster which is having minimum RS_RATIO.

var_clus_model.rsquare
```

```python
# Select the one variable from each cluster based on RS_RATIO
# RS_RATO = 1 - RS_Own_Cluster / 1 - RS_Next_Closest_Cluster
# Select one variable from each cluster which is having minimum RS_RATIO. 

var_clus = var_clus_model.rsquare
var_clus.to_csv("./var_clus.csv")
# var_clus.head(20)
```
